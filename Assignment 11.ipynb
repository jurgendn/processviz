**2.10.1**

Xác suất có 2 yêu cầu trong khoảng thời gian từ 0 đến 10 là:
$$
P(N(10) = 2) = \frac{0.3^2e^{-0.6}}{2!}
$$

**2.10.2**

Xác suất để yêu cầu đầu tiên xảy ra trước thời điểm 1 là:
$$
P(N(1) = 1|N(5) = 2) = P(N(1) = 1|N(5) - N(1) = 1) \\
= \frac{P(N(1) = 1)P(N(5) - N(1) = 1)}{P(N(5) - N(1) = 1)} \\
= P(N(1) = 1) = \frac{0.15^1e^{-0.15}}{1!}
$$

**2.10.3**

$N(t)$ là số lượng vụ tai nạn phát sinh tính đến thời điểm $t \to$ $N(t$)  là một quá trình Poisson với tham số $\lambda = 2$.

$X(t)$ là số lượng vụ tai nạn phát sinh vượt quá mức bảo hiểm cho đến thời điểm $t$ $\to X(t)$ là quá trình Poisson với tham số $\lambda_x = 0.25*2 = 0.5$.

Vậy xác suất để không có thiệt hại vượt quá mức bảo hiểm trong năm là 
$$
P(X(1) = 0) = \frac{0.5^1e^{-0.5*1}}{1}
$$

P(X \ge 60) = 1-P(X\le 60) \\

$$

**2.10.5**

Tham số $\lambda = 3$.

***a.*** Xác suất để từ 8h đến 12h không có bức điện nào là
$$
P(N(12) - N(8) = 0) = P(N(4) = 0)\\= \frac{e^{-3*4}}{0!} = ...
$$
***b.*** Gọi $X$ là thời điểm đầu tiên có thư sau $12h$.

Ta cần tính $P(X\ge x)$.  

Hiển nhiên $X = x$ là thời điểm đầu tiên nhận được bức điện sau 12h nên ta có $N(x) - N(12) = 0$.

Có
$$
P(X \ge x) = 1 - P(X \le x)\\=1 - P(N(x) - N(12) = 0) = 1 - \frac{\lambda^0.e^{\lambda (x-12)}}{0!} \\= 1 - e^{-\lambda (x - 12)}
$$
Do đó 
$$
Pmf(X = x) = \frac{d(P(X \le x))}{dx} \\= \lambda e^{\lambda(x-12)}
$$
