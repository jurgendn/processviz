Ta sẽ tính
$$
P(N(10) - N(0) = 2) = \frac{e^{-0.2*10}(0.2*10)^{2}}{2!} = 0.271
$$
**b.** Ta có $N(5) - N(0) = 2$. Ta cần tính $P(N(3)- N(0) = 1|N(5) - N(0) = 2)$

Ta có:
$$
\begin{aligned}
P(N(3) - N(0) \ge 1|N(5)-N(0) = 2) & = 1- P(N(3)<1|N(5)= 2) \\ 
&= 1-\frac{P(N(5)=2,N(3)<1)}{P(N(5)=2)} \\
& = 1-\sum_{k=0}^{1}\frac{P(N(5)=2,N(3)=k)}{P(N(5)=2)} \\
& = 1-\sum_{k=0}^{1}\frac{P(N(5) - N(3) = 2-k,N(3)=k)}{P(N(5)=2)} \\
& = 1 - 0.64 \\
& = 0.36
\end{aligned}
$$

$$
\begin{aligned}
P(X(1) = 2,X(3) = 6)  &= P(X(3) - X(1) = 4, X(1) = 2)\\
