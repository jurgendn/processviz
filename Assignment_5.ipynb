###### Định lí 1:

> Cho `C` là một `communicating class`(lớp liên thông). Khi đó tất cả các trạng thái trong `C` hoặc đều là `recurrent` hoặc đều là `transient`

***Chứng minh***

Lấy 2 trạng thái `i, j` bất kì trong `C`.

- Nếu `i` là transient. Dễ thấy tồn tại $m, n$ sao cho $p_{ij}^{(n)} > 0$ và $p_{ji}^{(m)}>0$.Ta có 
  $$
  p_{ii}^{(n+r+m)} \ge p_{ij}^{(n)}p_{jj}^{(r)}p_{ji}^{(m)}
  $$
  Do đó ta có
  $$
  \sum_{r=1}^{\infty}p_{jj}^{(r)} \ge \frac{1}{p_{ij}^{(n)}p_{ji}^{(m)}}\sum_{r = 1}^{\infty}p_{ii}^{(r)} < \infty
  $$
  Điều này chứng tỏ `j` là transient.

- Nếu `i` là recurrent, ta có:

  

$$
p_{jj}^{(n+r+m)} \ge p_{ji}^{(m)}p_{ii}^{(r)}p_{ij}^{(n)}
$$

Do `i` là recurrent nên `j` cũng là recurrent.

Ta có điều cần chứng minh.

###### Định lí 2:

> Cho `C` là một `communicating class` gồm các trạng thái $v_1, v_2,...,v_n$  Tập các đỉnh kề trong biểu diễn biểu diễn đồ thị chuyển trạng thái của các đỉnh $v_1, v_2,..,v_n$ lần lượt là $E_1, E_2, ...,E_n$.  Khi đó `C` là lớp đóng khi và chỉ khi $C = \cup_{i=1}^{n}E_i$

***Chứng minh***

`C` đóng $\Rightarrow$  $C = \cup_{i=1}^{n}E_i$.

Phản chứng, giả sử $C \neq \cup_{i=1}^{n}E_i$. Dễ thấy rằng $C \subset \cup_{i=1}^{n}E_i$. Do đó thấy rằng tồn tại một phần tử $t$ nào đó sao cho $t \in \cup_{i=1}^{n}E_i$ và $t \notin C$.

Chứng tỏ tồn tại một tập $E_i$ nào đó chứa $t$, dẫn đến đỉnh $v_i$ kề với $i$. Do $v_i \in C$ và $C$ đóng nên $t \in C$, vô lí. 

$\displaystyle C = \cup_{i=1}^{n}\{E_i\} \Rightarrow$ `C` đóng.

Giả sử $C$ không phải là lớp đóng. Khi đó tồn tại một đỉnh $t$ nào đó kề với một đỉnh $v_i \in C$ và $t \notin C$. Điều này vô lí với giả thiết ban đầu.

Vậy ra có điều cần chứng minh

###### Định lí 3:

> Mọi trạng thái trong một lớp liên thông đóng hữu hạn trạng thái đều là `recurrent`

***Chứng minh***

Giả sử tồn tại trạng thái $i \in C$ là `transient`. Khi đó tồn tại $n^*$ là số tự nhiên lớn nhất sao cho $p_{ii}^{n^*} > 0$. Ta chứng minh tồn tại số lớn hơn thoả mãn tính chất này.

Thật vậy, xét $p_{ii}^n$. Do $C$ là `irreducible`, ta đặt $n_{ij}: p_{ij}^{n_{ij}} > 0$. Ta có:

$$
p_{ij}^{n} \ge p_{ii_1}^{m_1}p_{i_1i_2}^{m_2}...p_{i_ni}^{m_{n+1}}, n = \sum_{k =1}^{n+1}m_{k}
$$

Khi đó, với mỗi $i_k \rightarrow i_{k+1}$, ta có thể lấy đoạn đường đi qua các đỉnh con $i_{k1}, i_{k2},..,i_{kn}, i_{k+1}$, mỗi đoạn này là tồn tại do xích là `irreducible`.

Quá trình này tiếp tục cho đến khi $n > n^*$, điều này trái với giả thiết ban đầu.

Điều vô lí này chứng tỏ các trạng thái của $C$ là `recurrent`

| $n$   | 0     | 0     | 0     | 0     | ...  | 1         | 0     |

Gọi $k$ là chỉ số lớn nhất sao cho $p_k > 0, k = \overline{0,n}, n \to \infty$

Khi đó ta sẽ chứng minh `R = {0,1,...,k}` là recurrent và `T = {k+1.k+2,...,}` là transient.

Thật vậy, dễ thấy rằng mọi $i < j \in R$  ta chứng minh tồn tại $n$ sao cho $p_{ij}^{(n)} > 0$ có: 
$$
p_{ij}^{(n)} = \Pi_{r=i}^{(1)}p_{i, i-1}.p_{k}.\Pi_{q = k}^{j+1}p_{q, q-1} > 0
$$
Do đó `R` là một `communicating class`. 

Mặt khác, dễ thấy `R` đóng và hữu hạn trạng thái, do đó các trạng thái trong `R` đều là recurrent.

Tiếp theo ta chứng minh các trạng thái trong`T` đều là transient. Thật vậy, ta sẽ chứng minh $p_{ii}^{(n)}=0, \forall n \ge 2, i \in T$

Ta có nhận xét sau:
$$
\left\{\begin{matrix}p_{ki} = 0, \forall k \in R, i \in T \\p_{ij} = 0, \forall i \le j \in T\end{matrix}\right.
$$
Do đó 
$$
p_{ii}^{(n)} = \sum_{r \in R}p_{ir}^{(n-1)}p_{ri} + \sum_{t \in T}p_{it}^{(n-1)}p_{ti} = 0
$$
Điều này chứng tỏ `T` là tập các trạng thái transient và `R` là tập các trạng thái recurrent.

***a.*** Do $(Y_n)$ đến được trạng thái $k$ sau $n$ bước nên phải tại 2 số tự nhiên $u, v$ sao cho
$$
\left\{\begin{matrix}u+v = n\\u-v = k\end{matrix}\right.
$$
Điều này chứng tỏ $2|n+k$.

Do đó nếu $2\nmid n+k$ thì $P(Y_n = k) = 0$.

Nếu $2|n+k$ thì khi đó, một bước đi từ 0 đến $k$ sau $n$ bước là cách sắp xếp $u$ bước +1 và $v$ bước -1. Số bước đi có thể là
$$
N = \binom{u+v}{u}
$$
do đó  $P(Y_n = k)$ được tính như sau
$$
\begin{aligned}P(Y_n = k) & = \binom{u+v}{u}\left(\frac{1}{2}\right)^u\left(\frac{1}{2}\right)^v \\& = \binom{n}{\frac{n+k}{2}}\left(\frac{1}{2}\right)^{n}\end{aligned}
$$
***b.*** Tại một thời điểm $n$ nào đó, $Y_n$ chỉ có thể thuộc vào 1 trong 2 khoảng $(-\infty, 0)$ và $(0, \infty)$.

Do đó
$$
P(Y_n < 0) = P(Y_n > 0) \Rightarrow P(Y_n \le 0) = P(Y_n \ge 0) = \frac{1}{2}
$$
***c.*** Có
$$
P(Y_n \le 2) = P(Y_n \le 0) + P(Y_n = 1) + P(Y_n = 2)
$$
Ta xét 2 trường hợp của $n$.

- $n = 0$. Dễ thấy $P(Y_n \le 2) = 1$

- $2|n$. Khi đó dễ thấy rằng $P(Y_n = 1) = 0$.
  Theo công thức đã chứng minh ở phần ***a,*** ta có:
  $$
  P(Y_n = 2) = \binom{n}{2}2^{-n}
  $$
  Vậy ta có:
  $$
  P(Y_n \le 2) = \frac{1}{2} + \frac{n(n-1)}{2^{n+1}}
  $$

- $2 \nmid n$. Khi đó dễ thấy rằng $P(Y_n = 2) = 0$
  Theo phần ***a,*** ta có:
  $$
  P(Y_n = 1) = \binom{n}{1}2^{-n}
  $$
  Từ đó ta có
  $$
  P(Y_n \le 2) = \frac{1}{2} + \frac{n}{2^{n}}
  $$

***Chứng minh $(Z_n)$ là xích Markov***.

Trước tiên ta chứng minh $(Y_n)$ là một xích Markov. Thật vậy, ta có:
$$
Y_n = \sum_{i=0}^{n}X_i = \sum_{i=0}^{n-1}X_i + X_n = Y_{n-1} + X_n
$$
Do đó ta có:
$$
\begin{aligned}P(Y_{n+1} = y_{n+1}|Y_{k} = y_k, k \le n) & = P(X_n = y_{n+1} - y_n|Y_{k} = y_k, k \le n-1) \\& =P(X_n = Y_{n+1} - y_n|Y_k = y_k, k \le n-1)\end{aligned}
$$

Hiển nhiên rằng $X_n$ độc lập với $S_k, k \le n-1$, do đó
$$
P(Y_{n+1} = y_{n+1}|Y_{k} = y_k, k \le n-1) = P(X_{n+1} = y_{n+1} - y_{n})
$$
Mặt khác, ta cũng có
$$
P(Y_{n+1} = y_{n+1}|Y_n = y_n) = P(X_{n+1} = y_{n+1} - y_n|Y_{n} = y_n) = P(X_{n+1} = y_{n+1}- y_n)
$$
Từ đó ta có $(Y_n)$ là xích Markov.

Có:
$$
\begin{aligned}P(Z_{n+1} = z_{n+1}|Z_{k} = z_{k}, k \le n) & = P(Y_{2n+2} = y_{2n+2}|Y_{2k} = y_{2k}, k \le n) \\& = P(X_{2n+1} + X_{2n+2} = y_{2n+2} - y_{2n}|Y_{2k} = y_{2k}, k\le n) \\& = P(X_{2n+2} + X_{2n+1} = y_{2n+2} - y_{2n}) \text{ do $X_{2n+2} + X_{2n+1} $ độc lập với $Y_n$}\end{aligned}
$$
Tương tự ta cũng chứng minh được
$$
P(Z_{n+1} = z_{n+1}|Z_n = z_n) = P(X_{2n+2} + X_{2n+1} = y_{2n+2} - y_{2n})
$$
Chứng tỏ $(Z_n)$ là một xích Markov.

Ta có $P(Z_{n+1} = z_{n+1}|Z_n = z_n) = P(X_{2n+2} + X_{2n+1} = y_{2n+2} - y_{2n})$. Mặt khác $X_i \in \{-1, ,\}$ nên dẫn đến $X_{2n+2} + X_{2n+1} \in \{-2,0,2\}$. Không gian trạng thái `I = {...,-2n,-2n+2n...,-2,0,2,...,2n-2,2n,...}`

\left\{\begin{matrix}P_{i, i-2} = p^2 \\P_{i, i} = 2pq \\P_{i, i+2} = q^2 \\\text{còn lại bằng 0}\end{matrix}\right. \forall i \in I

Chương trình minh họa như sau, source code đặt nằm trong `./processviz/algorithm/graph_travel.py`

```python
def get_communicating_class(state, P):
    connected_component = [[]]
    status = {i: False for i in state}
    while True:
        counter = 0
        for i in state:
            for j in state:
                if (BFS(state, P, i, j) and BFS(state, P, j, i)):
                    if status[i] == False:
                        connected_component[counter].append(i)
                        status[i] = True
                    if status[j] == False:
                        connected_component[counter].append(j)
                        status[j] = True
            connected_component.append([])
            counter += 1
        if i == state[len(state) - 1] and j == state[len(state) - 1]:
            break
    connected_component = list(filter(None, connected_component))
    return connected_component
```

Tiếp đó, để kiểm tra một trạng thái `state` nào đó là recurrent hay transient, trước tiên ta sẽ tìm một communicating class chứa nó. Nếu không có class nào chứa trạng thái này, hiển nhiên trạng thái này là `transient`

Chương trình như sau:

```python
def get_containing_class(s, P, state):
    communicating_class = get_communicating_class(s, P)
    for cl in communicating_class:
        if state in cl:
            return cl
    return False
```

Từ đây, sử dụng các định lí 2 và  3 ở phía trên, ta có thể kiểm tra tính hồi quy của một trạng thái như sau:

```python
def is_recurrent(self, state=None):
        if state != None:
            containing_class = gt.get_containing_class(
                self.state, self.P, state)
            if containing_class == False:
                return False
            else:
                state_vector = gt.convert_to_adjacency(self.state, self.P)
                union = set([])
                for i in containing_class:
                    union = union.union(set(state_vector[i]))
                return True if union == set(containing_class) else False
        else:
            return "No state is given"
```
